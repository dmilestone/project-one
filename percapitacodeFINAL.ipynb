```python
%matplotlib notebook
%matplotlib inline
import pandas as pd
import calendar
import datetime
import matplotlib.pyplot as plt
import numpy as np

# Study data files
popdata = "Resources/popdata.csv"
niche = "Resources/niche.csv"
crimedata = "Resources/COBRA-2009-2018.csv"

# Read the data
popdata = pd.read_csv(popdata)
niche = pd.read_csv(niche)
crimedata = pd.read_csv(crimedata)

crime_df = pd.DataFrame(crimedata)
crime_df.head()
```

```python
#creating new column singling out year of occurance
crime_df['year'] = pd.DatetimeIndex(crime_df['Occur Date']).year
crime_df.head()
```

```python
#seperating dataframes by year using loc
newdf2018 = crime_df.loc[crime_df["year"] == 2018,:]
newdf2017 = crime_df.loc[crime_df["year"] == 2017,:]
newdf2016 = crime_df.loc[crime_df["year"] == 2016,:]
newdf2015 = crime_df.loc[crime_df["year"] == 2015,:]
newdf2014 = crime_df.loc[crime_df["year"] == 2014,:]
```

```python
#this and the next 4 cells are getting incident counts by year for each neighborhood
necount2014 = newdf2014['Neighborhood'].value_counts()
crime2014_df = pd.DataFrame(necount2014)
newind2014 = crime2014_df.reset_index(drop=False)
fin2014 = newind2014.rename(columns={"Neighborhood":"2014 Count","index":"Neighborhood"})
fin2014.head()
```

```python
necount2015 = newdf2015['Neighborhood'].value_counts()
crime2015_df = pd.DataFrame(necount2015)
newind2015 = crime2015_df.reset_index(drop=False)
fin2015 = newind2015.rename(columns={"Neighborhood":"2015 Count","index":"Neighborhood"})
fin2015.head()
```

```python
necount2016 = newdf2016['Neighborhood'].value_counts()
crime2016_df = pd.DataFrame(necount2016)
newind2016 = crime2016_df.reset_index(drop=False)
fin2016 = newind2016.rename(columns={"Neighborhood":"2016 Count","index":"Neighborhood"})
fin2016.head()
```

```python
necount2017 = newdf2017['Neighborhood'].value_counts()
crime2017_df = pd.DataFrame(necount2017)
newind2017 = crime2017_df.reset_index(drop=False)
fin2017 = newind2017.rename(columns={"Neighborhood":"2017 Count","index":"Neighborhood"})
fin2017.head()
```

```python
necount2018 = newdf2018['Neighborhood'].value_counts()
crime2018_df = pd.DataFrame(necount2018)
newind2018 = crime2018_df.reset_index(drop=False)
fin2018 = newind2018.rename(columns={"Neighborhood":"2018 Count","index":"Neighborhood"})
fin2018.head()
```

```python
# merge population and niche price data (niche price=mediam home value from their website)
merged_df = pd.merge(niche, popdata, on="Neighborhood",how="left")
merged_df.head()
```

```python
###now merging my dataframes
second_df = pd.merge(merged_df, fin2014, on="Neighborhood",how="left")
third_df = pd.merge(second_df, fin2015, on="Neighborhood",how="left")
fourth_df = pd.merge(third_df, fin2016, on="Neighborhood",how="left")
fifth_df = pd.merge(fourth_df, fin2017, on="Neighborhood",how="left")
last_df = pd.merge(fifth_df, fin2018, on="Neighborhood",how="left")
last_df.head()
```

```python
###Crime Per Capita is typically displayed as incidents per 100,000 people, so multiplied per capita crime by that to get
#number of incidents per 100,000 people
last_df["2014 Crime Per Capita"] = (last_df["2014 Count"]/last_df["Population (2010)"])*100000
last_df["2015 Crime Per Capita"] = (last_df["2015 Count"]/last_df["Population (2010)"])*100000
last_df["2016 Crime Per Capita"] = (last_df["2016 Count"]/last_df["Population (2010)"])*100000
last_df["2017 Crime Per Capita"] = (last_df["2017 Count"]/last_df["Population (2010)"])*100000
last_df["2018 Crime Per Capita"] = (last_df["2018 Count"]/last_df["Population (2010)"])*100000
last_df.head()
```

```python
#saving to csv to analyze further
last_df.to_csv("Resources/niche_withcrimepercapita.csv")
```

```python
#Dropping null values since some of the neighborhoods were grouped together in 2010 population data and saving to csv
forgraphs_df = last_df.dropna()
forgraphs_df.to_csv("Resources/niche_withcrimepercapitawithdroppednulls.csv")
forgraphs_df.head()
```

```python
#highest priced neighborhoods (that had population data)
kingswood_df = forgraphs_df.loc[forgraphs_df["Neighborhood"] == "Kingswood",:]
argonne_df = forgraphs_df.loc[forgraphs_df["Neighborhood"] == "Argonne Forest",:]
margmitch_df = forgraphs_df.loc[forgraphs_df["Neighborhood"] == "Margaret Mitchell",:]
mtparannorthside_df = forgraphs_df.loc[forgraphs_df["Neighborhood"] == "Mt. Paran/Northside",:]
westpaces_df = forgraphs_df.loc[forgraphs_df["Neighborhood"] == "West Paces Ferry/Northside",:]

toconct = [kingswood_df,argonne_df,margmitch_df,mtparannorthside_df,westpaces_df]
foryrbyyr_df = pd.concat(toconct)
foryrbyyr_df
```

```python
#lowest priced neighborhoods (that had population data), also dropped outlier Marietta Street Artery with niche price $1194
carpark_df = forgraphs_df.loc[forgraphs_df["Neighborhood"] == "Carey Park",:]
almparke_df = forgraphs_df.loc[forgraphs_df["Neighborhood"] == "Almond Park",:]
pitts_df = forgraphs_df.loc[forgraphs_df["Neighborhood"] == "Pittsburgh",:]
hunthills_df = forgraphs_df.loc[forgraphs_df["Neighborhood"] == "Hunter Hills",:]
ashheight_df = forgraphs_df.loc[forgraphs_df["Neighborhood"] == "Ashview Heights",:]

toconct2 = [carpark_df,almparke_df,pitts_df,hunthills_df,ashheight_df]
lowestprice_df = pd.concat(toconct2)
lowestprice_df
```

```python
#Adding a column with the change in crime per capita over our time period
forchange_df = pd.DataFrame(forgraphs_df)
forchange_df["Change in Crime Per Capita"] = forchange_df["2018 Crime Per Capita"] - forchange_df["2014 Crime Per Capita"]
forchange_df.head()
```

```python
#Setting Niche Price as float for following graphs
def currency_to_float(value):
    no_dollar = value.replace("$", "")
    no_comma = no_dollar.replace(",", "")
    return float(no_comma)
forchange_df['Niche Price'] = forchange_df['Niche Price'].map(currency_to_float)
forchange_df.head()
```

```python
plt.scatter(forchange_df['Niche Price'],forchange_df['Change in Crime Per Capita'])
plt.title('Median Home Value vs. Change in Per Capita Crime 2014-2018')
plt.xlabel('Niche Price (Median Home Value)')
plt.ylabel('Change in Per Capita Crime from 2014-2018')
plt.savefig('scatterplot1')
plt.show()
```

```python
#Calculating the average crime per capita for all neighborhoods to compare against the top 5 and lowest 5 priced neighborhoods
avgpercap2014 = forchange_df["2014 Crime Per Capita"].mean()
avgpercap2015 = forchange_df["2015 Crime Per Capita"].mean()
avgpercap2016 = forchange_df["2016 Crime Per Capita"].mean()
avgpercap2017 = forchange_df["2017 Crime Per Capita"].mean()
avgpercap2018 = forchange_df["2018 Crime Per Capita"].mean()
avgpercap2014
```

```python
x_axis3 = [2014, 2015, 2016, 2017, 2018]
avgforall = [avgpercap2014,avgpercap2015,avgpercap2016,avgpercap2017,avgpercap2018]
topfivepercap2 = [foryrbyyr_df["2014 Crime Per Capita"],foryrbyyr_df["2015 Crime Per Capita"],foryrbyyr_df["2016 Crime Per Capita"],foryrbyyr_df["2017 Crime Per Capita"],foryrbyyr_df["2018 Crime Per Capita"]]


plt.suptitle("Crime Per Capita for Neighborhoods with Top 5 Median Home Value", fontsize=15)

plt.plot(x_axis3, topfivepercap2, linewidth=2, markersize=12)
plt.plot(x_axis3, avgforall, linewidth=4, markersize=12)
plt.xlabel('Year')
plt.ylabel('Crime Per Capita (per 100,000 people)')
plt.legend(['Kingswood','Argonne Forest','Margaret Mitchell','Mt. Paran/Northside','West Paces Ferry/Northside','Average for All Neighborhoods'], loc=0)
plt.savefig('top5line')
plt.show()
```

```python
x_axis4 = [2014, 2015, 2016, 2017, 2018]
lowfivepercap2 = [lowestprice_df["2014 Crime Per Capita"],lowestprice_df["2015 Crime Per Capita"],lowestprice_df["2016 Crime Per Capita"],lowestprice_df["2017 Crime Per Capita"],lowestprice_df["2018 Crime Per Capita"]]
#avgforall = [avgpercap2014,avgpercap2015,avgpercap2016,avgpercap2017,avgpercap2018]

plt.title("Crime Per Capita for Neighborhoods with 5 Lowest Median Home Values ", fontsize=15)

plt.plot(x_axis4, lowfivepercap2,linewidth=2, markersize=12)
plt.plot(x_axis4, avgforall, linewidth=4, markersize=12)

plt.xlabel('Year')
plt.ylabel('Crime Per Capita (per 100,000 people)')
plt.legend(['Carey Park','Almond Park','Pittsburgh','Hunter Hills','Ashview Heights','Average for All Neighborhoods'], loc=0)

plt.savefig('low5line')
plt.show()
```

```python
#Calculating the percentage of change in crime over each year
foryrbyyr_df["Total Change in Crime Per Capita"] = (foryrbyyr_df["2018 Crime Per Capita"] - foryrbyyr_df["2014 Crime Per Capita"])/foryrbyyr_df["2014 Crime Per Capita"]
foryrbyyr_df["2014-2015 Per Capita Change"] = (foryrbyyr_df["2015 Crime Per Capita"] - foryrbyyr_df["2014 Crime Per Capita"])/foryrbyyr_df["2014 Crime Per Capita"]
foryrbyyr_df["2015-2016 Per Capita Change"] = (foryrbyyr_df["2016 Crime Per Capita"] - foryrbyyr_df["2015 Crime Per Capita"])/foryrbyyr_df["2015 Crime Per Capita"]
foryrbyyr_df["2016-2017 Per Capita Change"] = (foryrbyyr_df["2017 Crime Per Capita"] - foryrbyyr_df["2016 Crime Per Capita"])/foryrbyyr_df["2016 Crime Per Capita"]
foryrbyyr_df["2017-2018 Per Capita Change"] = (foryrbyyr_df["2018 Crime Per Capita"] - foryrbyyr_df["2017 Crime Per Capita"])/foryrbyyr_df["2017 Crime Per Capita"]
foryrbyyr_df.head()
```
